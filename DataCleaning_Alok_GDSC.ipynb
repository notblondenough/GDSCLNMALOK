```python
import pandas as pd
```

```python
df = pd.read_csv('Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
df.isna().sum()
```

```python
#no data is missing so no need to fillna()
```

```python
# replace 0 values with column means
df['R&D Spend'] = df['R&D Spend'].replace(0, df['R&D Spend'].mean())
df['Administration'] = df['Administration'].replace(0, df['Administration'].mean())
df['Marketing Spend'] = df['Marketing Spend'].replace(0, df['Marketing Spend'].mean())
```

```python
df
```

```python
df.describe()
```

```python
#removing outliers in the given data
#using Z-Score Method
#R&D Spend
std=df['R&D Spend'].std()
mean=df['R&D Spend'].mean()
ul=mean+3*std;
ll=mean-3*std;
ul,ll
```

```python
df.loc[(df['R&D Spend']>ul)|(df['R&D Spend']<ll)]
```

```python
#all values lies in the range,no outlier detected
```

```python
#Administration
std=df['Administration'].std()
mean=df['Administration'].mean()
ul=mean+3*std;
ll=mean-3*std;
df.loc[(df['Administration']>ul)|(df['Administration']<ll)]
```

```python
#all values lies in the range,no outlier detected
```

```python
#Marketing Spend
std=df['Marketing Spend'].std()
mean=df['Marketing Spend'].mean()
ul=mean+3*std;
ll=mean-3*std;
df.loc[(df['Marketing Spend']>ul)|(df['Marketing Spend']<ll)]
```

```python
#all values lies in the range,no outlier detected
```

```python
#hence data cleaning is done on the given dataset
```
